# Data Analysis

```python
temp = []
for i in train.Activity:
    if i == "WALKING": temp.append(0)
    if i == "WALKING_UPSTAIRS": temp.append(1)
    if i == "WALKING_DOWNSTAIRS": temp.append(2)
    if i == "SITTING": temp.append(3)
    if i == "STANDING": temp.append(4)
    if i == "LAYING": temp.append(5)
train["n_Activity"] = temp
```

```python
temp = []
for i in test.Activity:
    if i == "WALKING": temp.append(0)
    if i == "WALKING_UPSTAIRS": temp.append(1)
    if i == "WALKING_DOWNSTAIRS": temp.append(2)
    if i == "SITTING": temp.append(3)
    if i == "STANDING": temp.append(4)
    if i == "LAYING": temp.append(5)
test["n_Activity"] = temp
```

```python
train.drop(["Activity"] , axis = 1 , inplace = True)
test.drop(["Activity"] , axis = 1 , inplace = True)
```

```python
df = train.copy()
label_1 = train["n_Activity"]
df.drop(["n_Activity"] , axis = 1 , inplace = True)
```

```python
df_test = test.copy()
#label_0 = test["n_Activity"]
df_test.drop(["n_Activity"] , axis = 1 , inplace = True)
```

# PCA AND t-SNE

```python
from sklearn.preprocessing import StandardScaler
standardized_data =  StandardScaler().fit_transform(df)
print(standardized_data.shape)
```

```python
sample_data = standardized_data

# matrix multiplication using numpy
covar_matrix = np.matmul(sample_data.T,sample_data)
print ( "The shape of variance matrix = ", covar_matrix.shape)
```

```python

from scipy.linalg import eigh

value,vector = eigh(covar_matrix,eigvals = (560,561))
print(vector.shape)


```

```python
vectors = vector.T
new_coordinates = np.matmul(vectors,sample_data.T)
print('the resultant new shape of the new coordinates', vectors.shape, "X", sample_data.T.shape ,'=', new_coordinates.shape)
```

```python
new_coordinates = np.vstack((new_coordinates, label_1)).T

# creating a new data frame for ploting the labeled points.
dataframe = pd.DataFrame(data=new_coordinates, columns=("1st_principal", "2nd_principal", "label_1"))
print(dataframe.shape)
print(dataframe.head())
```

```python
import seaborn as sn
sn.FacetGrid(dataframe, hue="label_1", height=5).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
plt.savefig("pca.png")
plt.show
```

```python
from sklearn.manifold import TSNE
modle = TSNE(n_components= 2,random_state = 0, perplexity=50, n_iter=5000)

# configuring the parameteres
# the number of components = 2
# default perplexity = 30
# default learning rate = 200
# default Maximum number of iterations for the optimization = 1000

tsne_data = modle.fit_transform(standardized_data)

# creating a new data frame which help us in ploting the result data
tsne_data = np.vstack((tsne_data.T,label_1)).T
tsne_df = pd.DataFrame(data = tsne_data,columns = ('dim1','dim2','labels'))
#ploting using seaborn
import seaborn as sns
sns.FacetGrid(tsne_df,hue = 'labels',size = 5).map(plt.scatter,'dim1','dim2').add_legend()
plt.savefig("tsne.png")
plt.show()
```

# MODEL FITTING

```python
from keras.utils import to_categorical
y_train = to_categorical(train.n_Activity , num_classes=6)
Y_test = to_categorical(test.n_Activity , num_classes=6)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,y_train,test_size=0.20,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train.iloc[:,0:562])
mat_train = scaler.transform(x_train.iloc[:,0:562])
print(mat_train)
```

```python
scaler = MinMaxScaler()
scaler.fit(df_test.iloc[:,0:562])
mat_test = scaler.transform(df_test.iloc[:,0:562])
print(mat_test)
```

```python
scaler = MinMaxScaler()
scaler.fit(x_test.iloc[:,0:562])
val_test = scaler.transform(x_test.iloc[:,0:562])
print(val_test)
```

```python
X_train = mat_train 
X_test = mat_test
X_val = val_test
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout , BatchNormalization
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
from keras.optimizers import RMSprop, Adam
```

```python
model = Sequential()

model.add(Dense(64, input_dim=X_train.shape[1] , activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu'))
model.add(Dense(196, activation='relu'))
model.add(Dropout(0.20))
model.add(Dense(32, activation='relu'))
model.add(Dense(6, activation='sigmoid'))

model.compile(optimizer = Adam(lr = 0.0001),loss='categorical_crossentropy', metrics=['accuracy'])
print(model.summary())
```

```python
history = model.fit(X_train, y_train , epochs=21 , batch_size = 64 , validation_data=(X_val, y_test))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python

from sklearn.metrics import confusion_matrix
pred = model.predict(X_test)
pred = np.argmax(pred,axis = 1) 
y_true = np.argmax(Y_test,axis = 1)
CM = confusion_matrix(y_true, pred)
from mlxtend.plotting import plot_confusion_matrix
fig, ax = plot_confusion_matrix(conf_mat=CM ,  figsize=(10, 5))
plt.savefig("confusion.png")
plt.show()
```

```python

from sklearn.metrics import classification_report , accuracy_score
print(classification_report(y_true, pred))
```

```python
print(accuracy_score(y_true, pred))
```
